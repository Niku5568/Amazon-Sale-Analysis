# Create Copy file of df for expand user id & user name columns

```python
df1=df.copy()
df2=df.copy()
df3=df.copy()
df4=df.copy()
df5=df.copy()
df6=df.copy()
df7=df.copy()
df8=df.copy()
```

```python
df1["user_id"]=df["user_id"].str.split(",",expand=True)[0]
df1["user_name"]=df["user_name"].str.split(",",expand=True)[0]
df2["user_id"]=df["user_id"].str.split(",",expand=True)[1]
df2["user_name"]=df["user_name"].str.split(",",expand=True)[1]
df3["user_id"]=df["user_id"].str.split(",",expand=True)[2]
df3["user_name"]=df["user_name"].str.split(",",expand=True)[2]
df4["user_id"]=df["user_id"].str.split(",",expand=True)[3]
df4["user_name"]=df["user_name"].str.split(",",expand=True)[3]
df5["user_id"]=df["user_id"].str.split(",",expand=True)[4]
df5["user_name"]=df["user_name"].str.split(",",expand=True)[4]
df6["user_id"]=df["user_id"].str.split(",",expand=True)[5]
df6["user_name"]=df["user_name"].str.split(",",expand=True)[5]
df7["user_id"]=df["user_id"].str.split(",",expand=True)[6]
df7["user_name"]=df["user_name"].str.split(",",expand=True)[6]
df8["user_id"]=df["user_id"].str.split(",",expand=True)[7]
df8["user_name"]=df["user_name"].str.split(",",expand=True)[7]

```

# Concate all 8 tables in 1 and store again in DF

```python
purchase=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8],ignore_index=True)
```

```python
sale=pd.read_csv("Sales.csv")
```

```python
df=df.drop_duplicates()
```

```python
df=df.drop(columns=["sub_category_x"])
```

```python
df=df.rename(columns={
    "categories_y":"categories",
    "product_type_y":"product_type",
    "rating_score_y":"rating_score",
    "Brand_y":"Brand",
    "sub_category_y":"sub_category"    
})
```

# top buyers that have sum of purchase amount higher

```python
highest_selling=df.groupby("user_name")[["selling_price"]].sum().reset_index().sort_values(by="selling_price",ascending=False).head(10)
highest_selling
```

```python
plt.figure(figsize=(8,6))
cbr=sns.barplot(data=highest_selling,x="user_name",y="selling_price")
plt.title("Sum of Selling Price of All Purchase")
plt.xlabel("Customer Name",fontweight='bold')
plt.ylabel("Sum of Selling Price",fontweight='bold')
plt.xticks(rotation=90)
for i in cbr.containers:
    cbr.bar_label(i, fmt='%.0f', fontweight='bold')
plt.grid()
plt.show()
```

# Brand with most number of products

```python
most_product=df.groupby("Brand")["Brand"].value_counts().sort_values(ascending=False).head(20).reset_index()
most_product
```

```python
plt.figure(figsize=(10,5))
mp=sns.barplot(most_product,x="Brand",y="count")
plt.title("Number of Product by Brand in Amazon")
plt.ylabel("Number of Products")
plt.xticks(rotation=90)
for i in mp.containers:
    mp.bar_label(i, fmt='%.0f', fontweight='bold')
```
