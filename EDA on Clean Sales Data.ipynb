# Values Distribution

```python
# Create histogram subplots on all numerical columns to understand distribustion 
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

sp=sns.histplot(df['selling_price'], kde=True,bins=10, ax=axes[0, 0])
axes[0, 0].set_title('Selling Price',fontweight='bold')
axes[0,0].set_xlabel("Selling Price", fontweight='bold')
axes[0,0].set_ylabel("Count", fontweight='bold')
for i in sp.containers:
    sp.bar_label(i, fmt='%.0f', fontweight='bold')

mrp=sns.histplot(df['MRP'], kde=True, bins=10, ax=axes[0, 1])
axes[0, 1].set_title('MRP',fontweight='bold')
axes[0,1].set_xlabel("MRP", fontweight='bold')
axes[0,1].set_ylabel("Count", fontweight='bold')
for i in mrp.containers:
    mrp.bar_label(i, fmt='%.0f', fontweight='bold')


dp=sns.histplot(df['discount_percentage'], kde=True,ax=axes[1, 0])
axes[1, 0].set_title('Discount Percentage',fontweight='bold')
axes[0,1].set_xlabel("Discount Percentage", fontweight='bold')
axes[0,1].set_ylabel("Count", fontweight='bold')
for i in dp.containers:
    dp.bar_label(i, fmt='%.0f', fontweight='bold')


rat=sns.histplot(df['rating'], kde=True, bins=10 ,ax=axes[1, 1])
axes[1, 1].set_title('Rating',fontweight='bold')
axes[0,1].set_xlabel("Rating", fontweight='bold')
axes[0,1].set_ylabel("Count", fontweight='bold')
for i in rat.containers:
    rat.bar_label(i, fmt='%.0f', fontweight='bold')


rc=sns.histplot(df['rating_count'], kde=True,bins=10, ax=axes[2, 0])
axes[2, 0].set_title('Rating Count',fontweight='bold')
axes[0,1].set_xlabel("Rating Count", fontweight='bold')
axes[0,1].set_ylabel("Count", fontweight='bold')
for i in rc.containers:
    rc.bar_label(i, fmt='%.0f', fontweight='bold')


pd=sns.histplot(df['Price_diffrence'], kde=True,bins=10, ax=axes[2, 1])
axes[2, 1].set_title('Price Difference',fontweight='bold')
axes[0,1].set_xlabel("Price Difference", fontweight='bold')
axes[0,1].set_ylabel("Count", fontweight='bold')
for i in pd.containers:
    pd.bar_label(i, fmt='%.0f', fontweight='bold')


plt.tight_layout()
plt.show()
```

># Distribution Observation: 
- Discount_percentage have sligthly normal distribution.
- column rating has positive skewness as tail are longer at right side.
- rest of other columns has negative skewness as tail are longer at left side.

> # Highest number of product in categories, sub-category and product_type 

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

cat_count=sns.countplot(ax=axes[0],y=df["categories"])
axes[0].set_title("Highest Number of Products in Catgories", fontweight='bold')
axes[0].set_ylabel("Category", fontweight='bold')
axes[0].set_xlabel("Count", fontweight='bold')
for i in cat_count.containers:
    cat_count.bar_label(i, fmt='%.0f', fontweight='bold')


sub_count=sns.countplot(ax=axes[1],y=df["sub_category"].head(600))
axes[1].set_title("Highest Number of Products in Sub Catgories", fontweight='bold')
axes[1].set_ylabel("Sub Category", fontweight='bold')
axes[1].set_xlabel("Count", fontweight='bold')
for i in sub_count.containers:
    sub_count.bar_label(i, fmt='%.0f', fontweight='bold')

pro_count=sns.countplot(ax=axes[2],y=df["product_type"].head(100))
axes[2].set_title("Highest Number of Products in product_type", fontweight='bold')
axes[2].set_ylabel("product_type", fontweight='bold')
axes[2].set_xlabel("Count", fontweight='bold')
for i in pro_count.containers:
    pro_count.bar_label(i, fmt='%.0f', fontweight='bold')

plt.tight_layout()
plt.show()
```

># Rating Score Distribution

```python
dist=df.groupby('rating_score').size().reset_index(name = "count")
dist
```

```python
rating_bar=sns.barplot(data=dist,x="rating_score",y="count",order=['Excellent','Good', 'Average','Below Average',"Poor"])
plt.title("Rating Score Distribution")
plt.xlabel("Rating Score")
for i in rating_bar.containers:
    rating_bar.bar_label(i, fmt='%.0f', fontweight='bold')

plt.show()
```
